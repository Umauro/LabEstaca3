### 5- Verosimilitud

#### Previo al Trabajo 

En esta sección se trabajará con el método de Newton-Raphson para encontrar el máximo estimador verosímil de $\lambda$ para una población que se distribuye como una Poisson con $\lambda$ desconocido ("poi.csv") y $\alpha$ para una distribución $Gamma(\alpha,\beta = 1)$ ("gamma.csv").

El método de Newton-Raphson consiste en lo siguiente:
\begin{aligned}
\theta _{n+1} = \theta _{n} - \frac{l'(\theta _{n})}{l''(\theta _{n})}
\end{aligned}
Donde: $\theta$ corresponde a lo que queremos estimar ($\lambda$ y $\alpha$) y $l'(\theta),l''(\theta)$ corresponden a la primera y segunda derivada de la función log-verosimilitud de $\theta$. El caso base $\theta _{0}$ es un estimador base arbitrario que elige uno mismo.

---

#### Poisson

Se sabe que la función de distribución de Poisson es:
\begin{aligned}
\frac{\lambda ^{x}\cdot e^{-\lambda}}{x!}
\end{aligned}
Aplicando $ln$ y luego derivando (desarrollo omitido) se obtiene la primera derivada de la función log-verosimilitud.
\begin{aligned}
l'(\lambda) = \frac{\sum_{i=1}^N X_{i}}{\lambda}-N
\end{aligned} 
Volviendo a derivar se obtiene la segunda derivada de la función log-verosimilitud. 
\begin{aligned}
l''(\lambda) = -\frac{\sum_{i=1}^N X_{i}}{\lambda ^{2}}
\end{aligned} 

Como se puede ver de $l'(\lambda)$ el estimador corresponde al promedio (despejando $\lambda$ al igualar a cero). Esto puede servir para comparar el resultado obtenido con el esperado.

---

#### Gamma

Se sabe que la función de distribución de Gamma con $\beta = 1$ es:
\begin{aligned}
\frac{1}{\Gamma (\alpha)}\cdot (X_{i})^{\alpha - 1}\cdot e^{-X_{i}}
\end{aligned}
Aplicando $ln$ y luego derivando (desarrollo omitido) se obtiene la primera derivada de la función log-verosimilitud.
\begin{aligned}
l'(\alpha) = -N\cdot \frac{\Gamma '(\alpha)}{\Gamma (\alpha)}+\sum_{i=1}^N ln(X_{i})
\end{aligned}
Volviendo a derivar se obtiene la segunda derivada de la función log-verosimilitud.
\begin{aligned}
l''(\alpha) = -N\cdot \left( \frac{\Gamma '(\alpha)}{\Gamma (\alpha)} \right)'
\end{aligned}

Donde $ \frac{\Gamma '(\alpha)}{\Gamma (\alpha)} $ es conocida como digamma y $\left( \frac{\Gamma '(\alpha)}{\Gamma (\alpha)} \right)'$ es conocida como trigamma. Disponibles en R y en Scipy. 

---

#### Profundizando Newton-Raphson
La sucesión de Newton-Raphson tiene un fin y eso es cuando se alcanza una convergencia definida como $| \theta_{n+1} - \theta_{n}| < u$ donde $u$ es un umbral de convergencia, cuando la diferencia entre ambos estimadores sucesivos sea menor que ese umbral, en teoría, se encontró el estimador máximo verosímil. Entonces se tiene que se realizar:
\begin{aligned}
\theta _{n+1} = \theta _{n} - \frac{l'(\theta _{n})}{l''(\theta _{n})}
\end{aligned}
Hasta que $| \theta_{n+1} - \theta_{n}| < u$ y luego se procede a entregar el estimador en cuestión.
